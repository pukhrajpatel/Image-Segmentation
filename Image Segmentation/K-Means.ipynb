```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('1j.jpg')
img  = cv2.resize(img, (800,600))
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img1 = img/255.0
img1 = img.reshape(-1,1)
img1.shape
```

```python
from sklearn.cluster import KMeans
Model = KMeans(n_clusters = 8)
Model.fit(img1)
```

```python
Kmeans_Img = Model.cluster_centers_
Kmeans_Img = Kmeans_Img[Model.labels_]
Kmeans_Img = Kmeans_Img.reshape(img.shape)
#cv2.imshow('IMG',Kmeans_Img)
#cv2.waitKey(0)
plt.plot(figsize = (12,10))
plt.imshow(Kmeans_Img)
plt.axis('off')
plt.title('K = 8')
plt.savefig('8k.jpg')
```
